```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the optimization function f(x)
def f(x):
    """
    Function to be plotted.
    Args: x: A 4-dimensional vector.
    Returns: The value of the function at x.
    """
    return (100 * (x[0]**2 - x[1])**2 + (x[0] - 1)**2 + (x[2] - 1)**2 + 90 * (x[1]**3 - x[3]) + 10.1 * ((x[1] - 1)**2 + (x[3] - 1)**2) + 19.8 * (x[1] - 1) * (x[3] - 1))
```

```python
t_values = np.linspace(-2, 2, 200)
#t_values
```

```python
# Generate a random start point a in [0, 1]^4
a = np.random.rand(4) # `rand` generates numbers from a uniform distribution between 0 and 1

# Generate a random direction vector d
d = np.random.randn(4) # randn generates numbers from a normal distribution with a mean of 0 and a standard deviation of 1
d /= np.linalg.norm(d) # Normalizing the vector d, such that ||d|| = 1

# evaluation 'f(a + td)' for each of 100 t's from -2 to 2 for random a (uniform distribution between 0 and 1)
# and random d (from a normal distribution) such that ||d|| = 1
function_values = [f(a + t * d) for t in t_values]
#function_values
```

```python
# Plot the f(x) values against t
plt.plot(t_values, function_values, label='f(a + td)')
plt.xlabel('$t$')
plt.ylabel('$f(a + td)$')
plt.title('Plot of $f(a + td)$ along a random direction')
plt.grid(True)
plt.show()
```
